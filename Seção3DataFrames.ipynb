```python
df2 = pd.DataFrame({"A": ['A0', 'A1', 'A2', 'A3'],
                    "B": ['B0', 'B1', 'B2', 'B3'],
                    "C": ['C0', 'C1', 'C2', 'C3'],
                    "D": ['D0', 'D1', 'D2', 'D3']},
                    index = [0,1,2,3])
```

```python
df3 = pd.DataFrame({"A": ['A4', 'A5', 'A6', 'A7'],
                    "B": ['B4', 'B5', 'B6', 'B7'],
                    "C": ['C4', 'C5', 'C6', 'C7'],
                    "D": ['D4', 'D5', 'D6', 'D7']},
                    index = [4,5,6,7])
```

```python
df4 = pd.DataFrame({"A": ['A8', 'A9', 'A10', 'A11'],
                    "B": ['B8', 'B9', 'B10', 'B11'],
                    "C": ['C8', 'C9', 'C10', 'C11'],
                    "D": ['D8', 'D9', 'D10', 'D11']},
                    index = [8,9,10,11])
```

```python
df2
```

```python
df3
```

```python
df4
```

```python
#Primeiro vamos executar um concatenação de DataFrames do modo estrutural errado
```

```python
frames = [df2, df3, df4]
#Se observarmos o FRAME acima estamos gerando uma lista
#Conseguimos provar isso executando o comando type(frames)
```

```python
frames
```

```python
type(frames)
```

```python
#Temos um comando específico no pandas para podermos concatenar de modo correto as DataFrame
```

```python
framesCombinados = pd.concat(frames)
```

```python
framesCombinados
```

```python
#Agora sim tenho uma estrutura DataFrame
#Utilizando o comando Type vemos isso
```

```python
type(framesCombinados)
```

```python
#ADICIONANDO SUB-GRUPOS A DATAFRAMES CONCATENADOS 
```

```python
grupo = pd.concat([df2,df3,df4], keys=['F1','F2','F3'])
#Keys seriam como chaves de sub-grupos
```

```python
grupo
```

```python
grupo['A']
#Caso queira selecionar somente a coluna A de todos os grupos
```

```python
grupo.loc['F2']
#Aqui vou localizar somente os dados da chave F2
```
